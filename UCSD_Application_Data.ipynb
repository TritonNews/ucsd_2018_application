```python
# package imports
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#stacked bar chart packages
from bokeh.core.properties import value
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.models import HoverTool
```

```python
#output_file("bar_stacked.html")
output_notebook()
```

```python
# import racial demographics data
race_data = pd.read_csv('../sheets/demo_2018.csv')
race_data.columns = ['campus', 'african_american', 'american_indian', 'asian_american', 'chicano_latino', 'pacific_islander',
                     'white', 'missing']
race_data
```

```python
output_file("bar_stacked.html")

campus =  race_data.campus.tolist()
racial_cat = ['African American', 'American Indian', 'Asian American', 'Chicano_Latino', 'Pacific Islander',
              'White', 'Missing']
colors = ['#fd32e5', '#2aa8ae', '#e13d2b', '#e06281', '#1695a4', '#3ff982', '#c9d9d3']

data = {'campus' : race_data.campus.tolist(),
        'African American'   : race_data.african_american.tolist(),
        'American Indian'   : race_data.american_indian.tolist(),
        'Asian American'   : race_data.asian_american.tolist(),
        'Chicano_Latino': race_data.chicano_latino.tolist(),
        'Pacific Islander': race_data.pacific_islander.tolist(),
        'White': race_data.white.tolist(),
        'Missing': race_data.missing.tolist(),}

source = ColumnDataSource(data=data)

hover = HoverTool(
  tooltips=[
    ('African American','@{African American}'),
    ('American Indian','@{American Indian}'),
    ('Asian American','@{Asian American}'),
    ('Chicano/Latino','@Chicano_Latino'),
    ('Pacific Islander','@{Pacific Islander}'),
    ('White','@White'),
    ('Missing','@Missing')],
)

p = figure(x_range=campus, plot_height=700, plot_width=900, title="Campus Racial Breakdown",
           toolbar_location=None, tools=[hover])

p.vbar_stack(racial_cat, x='campus', width=0.8, color=colors, source=source,
             legend=[value(x) for x in racial_cat])

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "vertical"
```

```python
show(p)
```
